```python
# Función para limpiar outliers
def reemplazar_outliers(dataframe, columna, metodo='mediana'):
    df = dataframe.copy()
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    if metodo == 'mediana':
        mediana = df[columna].median()
        df[columna] = np.where((df[columna] < limite_inferior) | (df[columna] > limite_superior), mediana, df[columna])
    elif metodo == 'media':
        media = df[columna].mean()
        df[columna] = np.where((df[columna] < limite_inferior) | (df[columna] > limite_superior), media, df[columna])
    elif metodo == 'bigotes':
        df[columna] = np.where((df[columna] < limite_inferior), limite_inferior, df[columna])
        df[columna] = np.where((df[columna] > limite_superior), limite_superior, df[columna])
    return df
```
